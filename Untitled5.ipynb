```python
# Importando bibliotecas necessrias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.preprocessing import StandardScaler

# 1. Carregando a base de dados Iris
iris = load_iris()
X = iris.data  # Atributos (comprimento e largura das spalas e ptalas)
y = iris.target  # Classes (rtulos)

# 2. Padronizando os dados (importante para K-Means e PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Aplicando K-Means para agrupamento
kmeans = KMeans(n_clusters=3, random_state=42)  # 3 clusters (uma para cada classe de flor)
kmeans.fit(X_scaled)
labels = kmeans.labels_  # Rtulos dos clusters

# 4. Reduzindo a dimensionalidade com PCA (para visualizao)
pca = PCA(n_components=2)  # Reduzindo para 2 dimenses
X_pca = pca.fit_transform(X_scaled)

# 5. Visualizando os clusters em 2D
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k', s=100)
plt.title('Agrupamento com K-Means (PCA reduzido para 2D)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.colorbar(label='Cluster')
plt.show()

# 6. Avaliando o agrupamento
# Silhouette Score: quanto mais prximo de 1, melhor
silhouette_avg = silhouette_score(X_scaled, labels)
print(f"Silhouette Score: {silhouette_avg:.2f}")

# Davies-Bouldin Index: quanto mais prximo de 0, melhor
davies_bouldin = davies_bouldin_score(X_scaled, labels)
print(f"Davies-Bouldin Index: {davies_bouldin:.2f}")

# 7. Comparando com os rtulos reais (apenas para anlise)
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', marker='o', edgecolor='k', s=100)
plt.title('Rtulos Reais (PCA reduzido para 2D)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.colorbar(label='Classe Real')
plt.show()
```
